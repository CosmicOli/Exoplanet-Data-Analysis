```python
import numpy as np;
from matplotlib import pyplot as mpl;
import pandas as pd;

mpl.figure().set_figwidth(15);

data = pd.read_csv("Kepler74.txt", delim_whitespace = True, header = None, names = ["time", "normalisedIntensities", "errorOnNormalisedIntensities"]);
mpl.errorbar(data["time"], data["normalisedIntensities"], yerr = data["errorOnNormalisedIntensities"], fmt = ".");

mpl.xlabel("Time (Days)");
mpl.ylabel("Normalised Intensity (Parts Per Million)");
mpl.title("Observations of Kepler 74");

mpl.show();
```
